10) –ù–∞–ø–∏—Å–∞—Ç—å 4 —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ä–æ–∂–¥–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ —Ü–µ–ª—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ (–ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é –∏ —É–±—ã–≤–∞–Ω–∏—é), —Å–ª—É—á–∞–π–Ω–æ–≥–æ –∏ –ø–æ—á—Ç–∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é. –ö–∞–∂–¥–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–ª–∏–Ω—É —Å–ø–∏—Å–∫–∞ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞.


```python
import random  # –ò–º–ø–æ—Ä—Ç –º–æ–¥—É–ª—è random –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª

def sorted_list(length):
    """–ì–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ —Ü–µ–ª—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é."""
    return list(range(1, length + 1))  # –°–æ–∑–¥–∞–Ω–∏–µ —Å–ø–∏—Å–∫–∞ –æ—Ç 1 –¥–æ length –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ

def reverse_sorted_list(length):
    """–ì–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ —Ü–µ–ª—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ —É–±—ã–≤–∞–Ω–∏—é."""
    return list(range(length, 0, -1))  # –°–æ–∑–¥–∞–Ω–∏–µ —Å–ø–∏—Å–∫–∞ –æ—Ç length –¥–æ 1 –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ

def random_list(length):
    """–ì–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç —Å–ª—É—á–∞–π–Ω—ã–π —Å–ø–∏—Å–æ–∫ —Ü–µ–ª—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π."""
    return [random.randint(1, 100) for _ in range(length)]  # –°–æ–∑–¥–∞–Ω–∏–µ —Å–ø–∏—Å–∫–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª –æ—Ç 1 –¥–æ 100 –¥–ª–∏–Ω—ã length

def almost_sorted_list(length):
    """–ì–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç –ø–æ—á—Ç–∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ —Ü–µ–ª—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é."""
    lst = list(range(1, length + 1))  # –°–æ–∑–¥–∞–Ω–∏–µ —Å–ø–∏—Å–∫–∞ –æ—Ç 1 –¥–æ length –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ
    for _ in range(length // 10):  # –ü–µ—Ä–µ–º–µ—à–∏–≤–∞–Ω–∏–µ 10% —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞
        index1 = random.randint(0, length - 1)  # –í—ã–±–æ—Ä —Å–ª—É—á–∞–π–Ω–æ–≥–æ –∏–Ω–¥–µ–∫—Å–∞
        index2 = random.randint(0, length - 1)  # –í—ã–±–æ—Ä –µ—â–µ –æ–¥–Ω–æ–≥–æ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –∏–Ω–¥–µ–∫—Å–∞
        lst[index1], lst[index2] = lst[index2], lst[index1]  # –û–±–º–µ–Ω –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –º–µ–∂–¥—É —ç–ª–µ–º–µ–Ω—Ç–∞–º–∏ –ø–æ –≤—ã–±—Ä–∞–Ω–Ω—ã–º –∏–Ω–¥–µ–∫—Å–∞–º
    return lst

# –ü—Ä–∏–º–µ—Ä—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è:
print("–û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫:", sorted_list(10))
print("–û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≤ –æ–±—Ä–∞—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ —Å–ø–∏—Å–æ–∫:", reverse_sorted_list(10))
print("–°–ª—É—á–∞–π–Ω—ã–π —Å–ø–∏—Å–æ–∫:", random_list(10))
print("–ü–æ—á—Ç–∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é —Å–ø–∏—Å–æ–∫:", almost_sorted_list(10))
```

11) –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏. –ü—Ä–æ—Ü–µ–¥—É—Ä–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –∞–ª–≥–æ—Ä–∏—Ç–º —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –∏ –∞–ª–≥–æ—Ä–∏—Ç–º –ø–æ—Ä–æ–∂–¥–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ f_gen(list_len), –≥–µ–Ω–µ—Ä–∏—Ä—É—é—â–∏–π —Å–ø–∏—Å–æ–∫ –∑–∞–¥–∞–Ω–Ω–æ–π –¥–ª–∏–Ω—ã, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–≤—Ç–æ—Ä–æ–µ–Ω–∏–π —Ç–µ—Å—Ç–∞ –∏ —Å–ø–∏—Å–æ–∫ –¥–ª–∏–Ω –∏—Å–ø–æ–ª—å–∑—É—é—â–∏—Ö—Å—è –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è. –î–ª—è –∫–∞–∂–¥–æ–π –¥–ª–∏–Ω—ã —Å–ø–∏—Å–∫–∞ —Ñ—É–Ω–∫—Ü–∏—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ, —Å—Ä–µ–¥–Ω–µ–µ –∏ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π –∏ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫.


```python
import random
import statistics

def test_sorting_algorithm(sort_algorithm, list_generator, repetitions, list_lengths):
    results = []  # –°–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è

    for length in list_lengths:  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –∫–∞–∂–¥–æ–π –¥–ª–∏–Ω–µ —Å–ø–∏—Å–∫–∞ –∏–∑ —Å–ø–∏—Å–∫–∞ list_lengths
        comparisons = []  # –°–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π
        permutations = []  # –°–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫

        for _ in range(repetitions):  # –ü–æ–≤—Ç–æ—Ä—è–µ–º —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∑–∞–¥–∞–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–∑
            test_list = list_generator(length)  # –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º —Å–ø–∏—Å–æ–∫ –∑–∞–¥–∞–Ω–Ω–æ–π –¥–ª–∏–Ω—ã
            comparisons_count, permutations_count = sort_algorithm(test_list)  # –ü—Ä–∏–º–µ–Ω—è–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –∫ —Å–ø–∏—Å–∫—É –∏ –ø–æ–ª—É—á–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π –∏ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫
            comparisons.append(comparisons_count)  # –î–æ–±–∞–≤–ª—è–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π –≤ —Å–ø–∏—Å–æ–∫ comparisons
            permutations.append(permutations_count)  # –î–æ–±–∞–≤–ª—è–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫ –≤ —Å–ø–∏—Å–æ–∫ permutations

        # –ù–∞—Ö–æ–¥–∏–º –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ, –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∏ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π
        min_comparisons = min(comparisons)
        max_comparisons = max(comparisons)
        avg_comparisons = statistics.mean(comparisons)

        # –ù–∞—Ö–æ–¥–∏–º –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ, –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∏ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫
        min_permutations = min(permutations)
        max_permutations = max(permutations)
        avg_permutations = statistics.mean(permutations)

        # –î–æ–±–∞–≤–ª—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –≤ —Å–ø–∏—Å–æ–∫ results
        results.append({
            'list_length': length,
            'min_comparisons': min_comparisons,
            'max_comparisons': max_comparisons,
            'avg_comparisons': avg_comparisons,
            'min_permutations': min_permutations,
            'max_permutations': max_permutations,
            'avg_permutations': avg_permutations
        })

    return results  # –í–æ–∑–≤—Ä–∞—â–∞–µ–º —Å–ø–∏—Å–æ–∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è

# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ test_sorting_algorithm:

# –ü—Ä–∏–º–µ—Ä –∞–ª–≥–æ—Ä–∏—Ç–º–∞ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ (–ø—É–∑—ã—Ä—å–∫–æ–≤–∞—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞)
def bubble_sort(arr):
    comparisons = 0  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—á–µ—Ç—á–∏–∫ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π
    permutations = 0  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—á–µ—Ç—á–∏–∫ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            comparisons += 1  # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π –ø—Ä–∏ –∫–∞–∂–¥–æ–º —Å—Ä–∞–≤–Ω–µ–Ω–∏–∏ —ç–ª–µ–º–µ–Ω—Ç–æ–≤
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                permutations += 1  # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫ –ø—Ä–∏ –∫–∞–∂–¥–æ–π –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–µ —ç–ª–µ–º–µ–Ω—Ç–æ–≤
    return comparisons, permutations  # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π –∏ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫

# –ü—Ä–∏–º–µ—Ä —Ñ—É–Ω–∫—Ü–∏–∏ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞
def generate_random_list(length):
    return [random.randint(0, 1000) for _ in range(length)]

repetitions = 10  # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π —Ç–µ—Å—Ç–∞
list_lengths = [10, 50, 100]  # –°–ø–∏—Å–æ–∫ –¥–ª–∏–Ω –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã—Ö –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è

# –í—ã–ø–æ–ª–Ω—è–µ–º —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∏ –≤—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
results = test_sorting_algorithm(bubble_sort, generate_random_list, repetitions, list_lengths)
for result in results:
    print(f"List length: {result['list_length']}, Min comparisons: {result['min_comparisons']}, Max comparisons: {result['max_comparisons']}, Avg comparisons: {result['avg_comparisons']}, Min permutations: {result['min_permutations']}, Max permutations: {result['max_permutations']}, Avg permutations: {result['avg_permutations']}")
```

12) –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –º–æ–¥–∏—Ñ–∏–∫–∞—Ü–∏—é —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –ø—É–∑—ã—Ä—å–∫–æ–º: "–∫–∞–º–µ—à–µ–∫-–ø—É–∑—ã—Ä–µ–∫" - —Å —á–µ—Ä–µ–¥—É—é—â–∏–º–∏—Å—è –ø—Ä–æ—Ö–æ–¥–∞–º–∏, –ø—Ä–∏–≤–æ–¥—è—â–∏–º–∏ –∫ "–≤—Å–ø–ª—ã—Ç–∏—é" —Å–∞–º–æ–≥–æ –±–æ–ª—å—à–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ –∏ "–ø–æ–≥—Ä—É–∂–µ–Ω–∏—é" —Å–∞–º–æ–≥–æ –º–∞–ª–µ–Ω—å–∫–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –∏–∑ –Ω–µ–æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —á–∞—Å—Ç–∏ —Å–ø–∏—Å–∫–∞.


```python
def stone_bubble_sort(arr):
    n = len(arr)  # –ü–æ–ª—É—á–∞–µ–º –¥–ª–∏–Ω—É —Å–ø–∏—Å–∫–∞
    for i in range(n):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ø–∏—Å–∫—É –æ—Ç 0 –¥–æ n-1
        # –í –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ —á–µ—Ç–Ω–æ–≥–æ –ø—Ä–æ—Ö–æ–¥–∞ (–≤—Å–ø–ª—ã—Ç–∏–µ)
        if i % 2 == 0:  # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á–µ—Ç–Ω–æ–µ –ª–∏ —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ i
            for j in range(0, n - i - 1, 2):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ø–∏—Å–∫—É —Å —à–∞–≥–æ–º 2
                if arr[j] > arr[j + 1]:  # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –Ω—É–∂–Ω–æ –ª–∏ –æ–±–º–µ–Ω—è—Ç—å —ç–ª–µ–º–µ–Ω—Ç—ã
                    arr[j], arr[j + 1] = arr[j + 1], arr[j]  # –û–±–º–µ–Ω–∏–≤–∞–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã
        # –í –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ –Ω–µ—á–µ—Ç–Ω–æ–≥–æ –ø—Ä–æ—Ö–æ–¥–∞ (–ø–æ–≥—Ä—É–∂–µ–Ω–∏–µ)
        else:  # –ï—Å–ª–∏ —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ i –Ω–µ—á–µ—Ç–Ω–æ–µ
            for j in range(1, n - i - 1, 2):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ø–∏—Å–∫—É —Å —à–∞–≥–æ–º 2, –Ω–∞—á–∏–Ω–∞—è —Å 1
                if arr[j] > arr[j + 1]:  # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –Ω—É–∂–Ω–æ –ª–∏ –æ–±–º–µ–Ω—è—Ç—å —ç–ª–µ–º–µ–Ω—Ç—ã
                    arr[j], arr[j + 1] = arr[j + 1], arr[j]  # –û–±–º–µ–Ω–∏–≤–∞–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã
    return arr  # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫

# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è:
arr = [64, 34, 25, 12, 22, 11, 90]  # –ó–∞–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
sorted_arr = stone_bubble_sort(arr)  # –í—ã–∑—ã–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
print("–û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –º–∞—Å—Å–∏–≤:", sorted_arr)  # –í—ã–≤–æ–¥–∏–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫
```

14) –ü—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º "–ø—É–∑—ã—Ä–µ–∫" –∏ "–∫–∞–º–µ—à–µ–∫-–ø—É–∑—ã—Ä–µ–∫" –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —Ç–∏–ø–∞—Ö —Å–ø–∏—Å–∫–æ–≤.


```python
def bubble_sort(arr):
    n = len(arr)  # –ü–æ–ª—É—á–∞–µ–º –¥–ª–∏–Ω—É —Å–ø–∏—Å–∫–∞
    for i in range(n):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ø–∏—Å–∫—É –æ—Ç 0 –¥–æ n-1
        for j in range(0, n - i - 1):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ø–∏—Å–∫—É –æ—Ç 0 –¥–æ n-i-1
            if arr[j] > arr[j + 1]:  # –ï—Å–ª–∏ —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç –±–æ–ª—å—à–µ —Å–ª–µ–¥—É—é—â–µ–≥–æ
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # –ú–µ–Ω—è–µ–º –∏—Ö –º–µ—Å—Ç–∞–º–∏
    return arr  # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫

def stone_bubble_sort(arr):
    n = len(arr)  # –ü–æ–ª—É—á–∞–µ–º –¥–ª–∏–Ω—É —Å–ø–∏—Å–∫–∞
    for i in range(n):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ø–∏—Å–∫—É –æ—Ç 0 –¥–æ n-1
        if i % 2 == 0:  # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á–µ—Ç–Ω–æ–µ –ª–∏ —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ i
            for j in range(0, n - i - 1, 2):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ø–∏—Å–∫—É —Å —à–∞–≥–æ–º 2
                if arr[j] > arr[j + 1]:  # –ï—Å–ª–∏ —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç –±–æ–ª—å—à–µ —Å–ª–µ–¥—É—é—â–µ–≥–æ
                    arr[j], arr[j + 1] = arr[j + 1], arr[j]  # –ú–µ–Ω—è–µ–º –∏—Ö –º–µ—Å—Ç–∞–º–∏
        else:  # –ï—Å–ª–∏ —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ i –Ω–µ—á–µ—Ç–Ω–æ–µ
            for j in range(1, n - i - 1, 2):  # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ø–∏—Å–∫—É —Å —à–∞–≥–æ–º 2, –Ω–∞—á–∏–Ω–∞—è —Å 1
                if arr[j] > arr[j + 1]:  # –ï—Å–ª–∏ —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç –±–æ–ª—å—à–µ —Å–ª–µ–¥—É—é—â–µ–≥–æ
                    arr[j], arr[j + 1] = arr[j + 1], arr[j]  # –ú–µ–Ω—è–µ–º –∏—Ö –º–µ—Å—Ç–∞–º–∏
    return arr  # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫

def test_sorting_algorithm(algorithm):
    # –¢–µ—Å—Ç–æ–≤—ã–µ —Å–ª—É—á–∞–∏ —Ä–∞–∑–Ω—ã—Ö —Ç–∏–ø–æ–≤ —Å–ø–∏—Å–∫–æ–≤
    test_cases = [
        [64, 34, 25, 12, 22, 11, 90],
        ["apple", "banana", "cherry", "date", "fig"],
        [5, "apple", 3.14, "banana", 1]
    ]
    # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤—Å–µ–º —Ç–µ—Å—Ç–æ–≤—ã–º —Å–ª—É—á–∞—è–º
    for case in test_cases:
        # –ü—Ä–∏–º–µ–Ω—è–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –∫ –∫–æ–ø–∏–∏ —Ç–µ—Å—Ç–æ–≤–æ–≥–æ —Å–ª—É—á–∞—è
        sorted_case = algorithm(case.copy())
        # –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
        print(f"–°–ø–∏—Å–æ–∫ –¥–æ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏: {case}")
        print(f"–°–ø–∏—Å–æ–∫ –ø–æ—Å–ª–µ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏: {sorted_case}")
        print()

print("–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø—É–∑—ã—Ä—å–∫–æ–º:")
test_sorting_algorithm(bubble_sort)
print("–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –∫–∞–º–µ—à–µ–∫-–ø—É–∑—ã—Ä—å–∫–æ–º:")
test_sorting_algorithm(stone_bubble_sort)
```

15) –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –º–æ–¥–∏—Ñ–∏–∫–∞—Ü–∏—é —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –≤—ã–±–æ—Ä–æ–º –Ω–∞ –æ—Å–Ω–æ–≤–µ –≤—ã–±–æ—Ä–∞ –∏–∑ –ø–æ–¥—Å–ø–∏—Å–∫–æ–≤ (–æ–ø–∏—Å–∞–Ω–Ω–æ–≥–æ –≤ –ª–µ–∫—Ü–∏–∏). –ü—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º.


```python
def selection_sort_modified(arr):
    # –í–Ω—É—Ç—Ä–µ–Ω–Ω—è—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–∏—Å–∫–∞ –∏–Ω–¥–µ–∫—Å–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ –ø–æ–¥—Å–ø–∏—Å–∫–µ
    def find_min_index(arr, start):
        min_index = start
        # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤—Å–µ–º —ç–ª–µ–º–µ–Ω—Ç–∞–º –ø–æ—Å–ª–µ —Å—Ç–∞—Ä—Ç–æ–≤–æ–≥–æ –∏–Ω–¥–µ–∫—Å–∞ –∏ –∏—â–µ–º –º–∏–Ω–∏–º—É–º
        for i in range(start + 1, len(arr)):
            if arr[i] < arr[min_index]:
                min_index = i
        return min_index

    # –û—Å–Ω–æ–≤–Ω–æ–π —Ü–∏–∫–ª —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –≤—ã–±–æ—Ä–æ–º
    for i in range(0, len(arr) - 1):
        # –ù–∞—Ö–æ–¥–∏–º –∏–Ω–¥–µ–∫—Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ —Ç–µ–∫—É—â–µ–º –ø–æ–¥—Å–ø–∏—Å–∫–µ
        min_index = find_min_index(arr, i)
        # –ï—Å–ª–∏ –∏–Ω–¥–µ–∫—Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –Ω–µ —Ä–∞–≤–µ–Ω —Ç–µ–∫—É—â–µ–º—É –∏–Ω–¥–µ–∫—Å—É, –º–µ–Ω—è–µ–º –∏—Ö –º–µ—Å—Ç–∞–º–∏
        if min_index != i:
            arr[i], arr[min_index] = arr[min_index], arr[i]

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –º–æ–¥–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –≤—ã–±–æ—Ä–æ–º
def test_selection_sort_modified():
    # –¢–µ—Å—Ç–æ–≤—ã–µ —Å–ª—É—á–∞–∏
    test_cases = [
        [],
        [1],
        [3, 2, 1],
        [5, 2, 7, 3, 9, 1],
        [1, 2, 3, 4, 5],
        [5, 4, 3, 2, 1]
    ]

    # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –∫–∞–∂–¥–æ–º—É —Ç–µ—Å—Ç–æ–≤–æ–º—É —Å–ª—É—á–∞—é
    for i, test_case in enumerate(test_cases):
        original = test_case.copy()
        # –°–æ—Ä—Ç–∏—Ä—É–µ–º –º–∞—Å—Å–∏–≤ —Å –ø–æ–º–æ—â—å—é –º–æ–¥–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –≤—ã–±–æ—Ä–æ–º
        selection_sort_modified(test_case)
        print(f"Test case {i+1}:")
        print(f"Original array: {original}")
        print(f"Sorted array: {test_case}")
        print()

# –ü—Ä–æ–≤–µ—Ä—è–µ–º, –∑–∞–ø—É—â–µ–Ω –ª–∏ —Å–∫—Ä–∏–ø—Ç –Ω–∞–ø—Ä—è–º—É—é, –∏ –µ—Å–ª–∏ –¥–∞, —Ç–æ –∑–∞–ø—É—Å–∫–∞–µ–º —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ
if __name__ == "__main__":
    test_selection_sort_modified()
```

16) –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –±—ã—Å—Ç—Ä—É—é —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫—É —Å –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å—é –ø–æ–¥—Å—á–µ—Ç–∞ –æ–ø–µ—Ä–∞—Ü–∏–π. –°—Ä–∞–≤–Ω–∏—Ç—å –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –Ω–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö –∏ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö (–∏ –ø–æ—á—Ç–∏ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö) –¥–∞–Ω–Ω—ã—Ö.


```python
def quicksort(arr, start=0, end=None, count_operations=False):
    """
    –†–µ–∞–ª–∏–∑–∞—Ü–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –±—ã—Å—Ç—Ä–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - arr: —Å–ø–∏—Å–æ–∫ –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏.
    - start: –Ω–∞—á–∞–ª—å–Ω—ã–π –∏–Ω–¥–µ–∫—Å –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 0).
    - end: –∫–æ–Ω–µ—á–Ω—ã–π –∏–Ω–¥–µ–∫—Å –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é None, —á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç –∫–æ–Ω–µ—Ü —Å–ø–∏—Å–∫–∞).
    - count_operations: —Ñ–ª–∞–≥ –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –æ–ø–µ—Ä–∞—Ü–∏–π (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é False).

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    –ù–∏—á–µ–≥–æ. –§—É–Ω–∫—Ü–∏—è —Å–æ—Ä—Ç–∏—Ä—É–µ—Ç —Å–ø–∏—Å–æ–∫ arr –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è.
    """

    # –ï—Å–ª–∏ –∫–æ–Ω–µ—á–Ω—ã–π –∏–Ω–¥–µ–∫—Å –Ω–µ —É–∫–∞–∑–∞–Ω, —É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –µ–≥–æ –Ω–∞ –∫–æ–Ω–µ—Ü —Å–ø–∏—Å–∫–∞
    if end is None:
        end = len(arr) - 1

    # –ï—Å–ª–∏ –Ω–∞—á–∞–ª—å–Ω—ã–π –∏–Ω–¥–µ–∫—Å –º–µ–Ω—å—à–µ –∫–æ–Ω–µ—á–Ω–æ–≥–æ, –≤—ã–ø–æ–ª–Ω—è–µ–º —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫—É
    if start < end:
        # –ù–∞—Ö–æ–¥–∏–º –∏–Ω–¥–µ–∫—Å –æ–ø–æ—Ä–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –ø–æ—Å–ª–µ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞
        pivot_index = partition(arr, start, end, count_operations)
        # –†–µ–∫—É—Ä—Å–∏–≤–Ω–æ —Å–æ—Ä—Ç–∏—Ä—É–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã –ø–µ—Ä–µ–¥ –æ–ø–æ—Ä–Ω—ã–º –∏ –ø–æ—Å–ª–µ –Ω–µ–≥–æ
        quicksort(arr, start, pivot_index - 1, count_operations)
        quicksort(arr, pivot_index + 1, end, count_operations)


def partition(arr, start, end, count_operations):
    """
    –í—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –±—ã—Å—Ç—Ä–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏.
    –†–∞–∑–¥–µ–ª—è–µ—Ç —Å–ø–∏—Å–æ–∫ –Ω–∞ –¥–≤–µ —á–∞—Å—Ç–∏: —ç–ª–µ–º–µ–Ω—Ç—ã –º–µ–Ω—å—à–µ –æ–ø–æ—Ä–Ω–æ–≥–æ –∏ —ç–ª–µ–º–µ–Ω—Ç—ã –±–æ–ª—å—à–µ –æ–ø–æ—Ä–Ω–æ–≥–æ.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - arr: —Å–ø–∏—Å–æ–∫ –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏.
    - start: –Ω–∞—á–∞–ª—å–Ω—ã–π –∏–Ω–¥–µ–∫—Å –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è.
    - end: –∫–æ–Ω–µ—á–Ω—ã–π –∏–Ω–¥–µ–∫—Å –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è.
    - count_operations: —Ñ–ª–∞–≥ –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –æ–ø–µ—Ä–∞—Ü–∏–π.

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    –ò–Ω–¥–µ–∫—Å –æ–ø–æ—Ä–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –ø–æ—Å–ª–µ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞.
    """

    # –ó–∞–¥–∞–µ–º –æ–ø–æ—Ä–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç (–≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ - –ø–æ—Å–ª–µ–¥–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç —Å–ø–∏—Å–∫–∞)
    pivot = arr[end]
    i = start - 1
    # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ø–∏—Å–∫—É –∏ –ø–µ—Ä–µ–º–µ—â–∞–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã –º–µ–Ω—å—à–µ –æ–ø–æ—Ä–Ω–æ–≥–æ –≤–ª–µ–≤–æ –æ—Ç –Ω–µ–≥–æ
    for j in range(start, end):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
            # –ï—Å–ª–∏ –≤–∫–ª—é—á–µ–Ω –ø–æ–¥—Å—á–µ—Ç –æ–ø–µ—Ä–∞—Ü–∏–π, —É–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫ –Ω–∞ 1
            if count_operations:
                count_operations[0] += 1  # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–µ—Ä–∞—Ü–∏–π –Ω–∞ 1
    # –ü–µ—Ä–µ–º–µ—â–∞–µ–º –æ–ø–æ—Ä–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –≤ –ø—Ä–∞–≤–∏–ª—å–Ω—É—é –ø–æ–∑–∏—Ü–∏—é
    arr[i + 1], arr[end] = arr[end], arr[i + 1]
    if count_operations:
        count_operations[0] += 1  # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–µ—Ä–∞—Ü–∏–π –Ω–∞ 1
    return i + 1


def test_quicksort_performance(data, ordered=False):
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –±—ã—Å—Ç—Ä–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –Ω–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö –∏–ª–∏ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - data: —Å–ø–∏—Å–æ–∫ –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏.
    - ordered: —Ñ–ª–∞–≥, —É–∫–∞–∑—ã–≤–∞—é—â–∏–π, —è–≤–ª—è—é—Ç—Å—è –ª–∏ –¥–∞–Ω–Ω—ã–µ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã–º–∏ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é False).

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    –ö–æ—Ä—Ç–µ–∂ (–≤—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–µ—Ä–∞—Ü–∏–π).
    """

    import time
    import random

    # –ï—Å–ª–∏ –¥–∞–Ω–Ω—ã–µ –Ω–µ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω—ã –∏ ordered=True, —Å–æ—Ä—Ç–∏—Ä—É–µ–º –∏—Ö –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é
    if ordered:
        data.sort()

    # –ó–∞—Å–µ–∫–∞–µ–º –≤—Ä–µ–º—è –Ω–∞—á–∞–ª–∞ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è
    start_time = time.time()
    # –°–æ–∑–¥–∞–µ–º —Å—á–µ—Ç—á–∏–∫ –æ–ø–µ—Ä–∞—Ü–∏–π
    count_operations = [0]  # –°—á–µ—Ç—á–∏–∫ –æ–ø–µ—Ä–∞—Ü–∏–π
    # –í—ã–ø–æ–ª–Ω—è–µ–º –±—ã—Å—Ç—Ä—É—é —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫—É
    quicksort(data, count_operations=count_operations)
    # –ó–∞—Å–µ–∫–∞–µ–º –≤—Ä–µ–º—è –æ–∫–æ–Ω—á–∞–Ω–∏—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è
    end_time = time.time()

    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –≤—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–µ—Ä–∞—Ü–∏–π
    return end_time - start_time, count_operations[0]


if __name__ == "__main__":
    # –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º —Å–ª—É—á–∞–π–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
    random_data = [random.randint(0, 1000) for _ in range(1000)]

    # –¢–µ—Å—Ç–∏—Ä—É–µ–º –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –Ω–∞ —Å–ª—É—á–∞–π–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    random_time, random_operations = test_quicksort_performance(random_data.copy())
    print("Quicksort performance on random data:")
    print("Time:", random_time, "seconds")
    print("Number of operations:", random_operations)

    # –¢–µ—Å—Ç–∏—Ä—É–µ–º –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –Ω–∞ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    ordered_time, ordered_operations = test_quicksort_performance(random_data.copy(), ordered=True)
    print("\nQuicksort performance on ordered data:")
    print("Time:", ordered_time, "seconds")
    print("Number of operations:", ordered_operations)
```

17) –£–ª—É—á—à–∏—Ç—å —Ä–∞–±–æ—Ç—É –±—ã—Å—Ç—Ä–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –Ω–∞ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –∏ –ø–æ—á—Ç–∏ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –∑–∞ —Å—á–µ—Ç –∏–∑–º–µ–Ω–µ–Ω–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –≤—ã–±–æ—Ä–∞ —ç–ª–µ–º–µ–Ω—Ç–∞ –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è –º–∞—Å—Å–∏–≤–∞.


```python
def partition(arr, low, high):
    # –§—É–Ω–∫—Ü–∏—è partition –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è –º–∞—Å—Å–∏–≤–∞ –Ω–∞ –¥–≤–µ —á–∞—Å—Ç–∏ –≤–æ–∫—Ä—É–≥ –æ–ø–æ—Ä–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞
    # –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∏–Ω–¥–µ–∫—Å –æ–ø–æ—Ä–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –ø–æ—Å–ª–µ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è.
    
    # –í—ã–±–∏—Ä–∞–µ–º –º–µ–¥–∏–∞–Ω—É –∏–∑ –ø–µ—Ä–≤–æ–≥–æ, —Å—Ä–µ–¥–Ω–µ–≥–æ –∏ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –æ–ø–æ—Ä–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞
    mid = (low + high) // 2
    pivot_index = high
    if arr[low] < arr[mid]:
        if arr[mid] < arr[high]:
            pivot_index = mid
    elif arr[low] < arr[high]:
        pivot_index = low

    pivot = arr[pivot_index]

    # –ü–µ—Ä–µ—Å—Ç–∞–≤–ª—è–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã —Ç–∞–∫, —á—Ç–æ–±—ã –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã —Å–ª–µ–≤–∞ –æ—Ç pivot –±—ã–ª–∏ –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω—ã –µ–º—É,
    # –∞ –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã —Å–ø—Ä–∞–≤–∞ - –±–æ–ª—å—à–µ –Ω–µ–≥–æ
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1


def quick_sort(arr, low, high):
    # –§—É–Ω–∫—Ü–∏—è quick_sort –≤—ã–ø–æ–ª–Ω—è–µ—Ç –±—ã—Å—Ç—Ä—É—é —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫—É –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ –∏–Ω–¥–µ–∫—Å–æ–≤ –º–∞—Å—Å–∏–≤–∞
    
    if low < high:
        # –ï—Å–ª–∏ –¥–∏–∞–ø–∞–∑–æ–Ω —Å–æ–¥–µ—Ä–∂–∏—Ç –±–æ–ª–µ–µ –æ–¥–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞, –ø—Ä–æ–∏–∑–≤–æ–¥–∏–º —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –∏ —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ —Å–æ—Ä—Ç–∏—Ä—É–µ–º
        # –∫–∞–∂–¥—É—é –ø–æ–ª–æ–≤–∏–Ω—É –º–∞—Å—Å–∏–≤–∞
        
        # –†–∞–∑–±–∏–≤–∞–µ–º –º–∞—Å—Å–∏–≤ –∏ —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ —Å–æ—Ä—Ç–∏—Ä—É–µ–º –∫–∞–∂–¥—É—é –µ–≥–æ –ø–æ–ª–æ–≤–∏–Ω—É
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)


# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è:
arr = [10, 7, 8, 9, 1, 5]
n = len(arr)
quick_sort(arr, 0, n - 1)
print("–û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –º–∞—Å—Å–∏–≤:", arr)
```

18) –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º –ø–æ–ª—É—á–µ–Ω–∏—è –∏–∑ –±–æ–ª—å—à–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —Å–ø–∏—Å–∫–æ–≤ –æ–¥–Ω–æ–≥–æ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞.


```python
import heapq  # –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –º–æ–¥—É–ª—å heapq –¥–ª—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –∫—É—á–∏

def merge_sorted_lists(sorted_lists):
    merged_list = []  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ –¥–ª—è –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω–æ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
    heap = []  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø—É—Å—Ç—É—é –∫—É—á—É

    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∫—É—á–∏
    for i, sublist in enumerate(sorted_lists):  # –ü–µ—Ä–µ–±–∏—Ä–∞–µ–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ —Å–ø–∏—Å–∫–∏ –∏ –∏—Ö –∏–Ω–¥–µ–∫—Å—ã
        if sublist:  # –ï—Å–ª–∏ —Å–ø–∏—Å–æ–∫ –Ω–µ –ø—É—Å—Ç–æ–π
            heapq.heappush(heap, (sublist[0], i, 0))  # –î–æ–±–∞–≤–ª—è–µ–º –≤ –∫—É—á—É –∫–æ—Ä—Ç–µ–∂ (–∑–Ω–∞—á–µ–Ω–∏–µ, –∏–Ω–¥–µ–∫—Å_—Å–ø–∏—Å–∫–∞, –∏–Ω–¥–µ–∫—Å_–∑–Ω–∞—á–µ–Ω–∏—è)

    while heap:  # –ü–æ–∫–∞ –∫—É—á–∞ –Ω–µ –ø—É—Å—Ç–∞
        val, list_idx, idx = heapq.heappop(heap)  # –ò–∑–≤–ª–µ–∫–∞–µ–º –Ω–∞–∏–º–µ–Ω—å—à–∏–π —ç–ª–µ–º–µ–Ω—Ç –∏–∑ –∫—É—á–∏
        merged_list.append(val)  # –î–æ–±–∞–≤–ª—è–µ–º –∏–∑–≤–ª–µ—á–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫

        if idx + 1 < len(sorted_lists[list_idx]):  # –ï—Å–ª–∏ –≤ —Å–ø–∏—Å–∫–µ –µ—â–µ –µ—Å—Ç—å —ç–ª–µ–º–µ–Ω—Ç—ã
            next_val = sorted_lists[list_idx][idx + 1]  # –ü–æ–ª—É—á–∞–µ–º —Å–ª–µ–¥—É—é—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ —Ç–æ–≥–æ –∂–µ —Å–ø–∏—Å–∫–∞
            heapq.heappush(heap, (next_val, list_idx, idx + 1))  # –î–æ–±–∞–≤–ª—è–µ–º –µ–≥–æ –≤ –∫—É—á—É

    return merged_list  # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫

# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
sorted_lists = [
    [1, 3, 5],
    [2, 4, 6],
    [0, 7, 8]
]
result = merge_sorted_lists(sorted_lists)
print(result)
```

19) –ú–æ–¥–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞—Ç—å –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –¥–ª–∏–Ω —à–∞–≥–æ–≤ –≤ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–µ –®–µ–ª–ª–∞.
–ù–∞¬†2[ùëõ/2ùëò+1]+1,—Ç.–µ. –ø—Ä–∏ –ø–µ—Ä–≤–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–µ –±—É–¥—É—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è —à–∞–≥–∏: 2[n/4]+1, 2[n/8]+1, ... , 3, 1 .

```python
def shell_sort(arr):
    n = len(arr)  # –ü–æ–ª—É—á–∞–µ–º –¥–ª–∏–Ω—É –º–∞—Å—Å–∏–≤–∞
    gap = n // 2  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —à–∞–≥ –∫–∞–∫ –ø–æ–ª–æ–≤–∏–Ω—É –¥–ª–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞
    while gap > 0:  # –ü–æ–∫–∞ —à–∞–≥ –±–æ–ª—å—à–µ –Ω—É–ª—è
        for i in range(gap, n):  # –ù–∞—á–∏–Ω–∞—è —Å —ç–ª–µ–º–µ–Ω—Ç–∞ —Å –∏–Ω–¥–µ–∫—Å–æ–º gap –∏ –¥–æ –∫–æ–Ω—Ü–∞ –º–∞—Å—Å–∏–≤–∞
            temp = arr[i]  # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç
            j = i
            while j >= gap and arr[j - gap] > temp:  # –ü–æ–∫–∞ –Ω–µ –¥–æ—Å—Ç–∏–≥–Ω—É—Ç –Ω–∞—á–∞–ª–æ –º–∞—Å—Å–∏–≤–∞ –∏ —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç –º–µ–Ω—å—à–µ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –Ω–∞ gap
                arr[j] = arr[j - gap]  # –ü–µ—Ä–µ–º–µ—â–∞–µ–º —ç–ª–µ–º–µ–Ω—Ç –≤–ø–µ—Ä–µ–¥ –Ω–∞ gap –ø–æ–∑–∏—Ü–∏–π
                j -= gap  # –£–º–µ–Ω—å—à–∞–µ–º –∏–Ω–¥–µ–∫—Å –Ω–∞ —à–∞–≥
            arr[j] = temp  # –í—Å—Ç–∞–≤–ª—è–µ–º —Å–æ—Ö—Ä–∞–Ω–µ–Ω–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –Ω–∞ –ø—Ä–∞–≤–∏–ª—å–Ω—É—é –ø–æ–∑–∏—Ü–∏—é
        gap = (gap * 2) // 2 + 1  # –ú–æ–¥–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω–∞—è —Ñ–æ—Ä–º—É–ª–∞ –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ —à–∞–≥–∞
    return arr  # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –º–∞—Å—Å–∏–≤

# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è:
arr = [12, 34, 54, 2, 3]
print("–ò—Å—Ö–æ–¥–Ω—ã–π –º–∞—Å—Å–∏–≤:", arr)
sorted_arr = shell_sort(arr)
print("–û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –º–∞—Å—Å–∏–≤:", sorted_arr)
```
