10) Написать 4 функции порождения списка целых значений: отсортированного (по возрастанию и убыванию), случайного и почти отсортированного по возрастанию. Каждая функция принимает длину списка в качестве аргумента.


```python
import random  # Импорт модуля random для генерации случайных чисел

def sorted_list(length):
    """Генерирует отсортированный список целых значений по возрастанию."""
    return list(range(1, length + 1))  # Создание списка от 1 до length включительно

def reverse_sorted_list(length):
    """Генерирует отсортированный список целых значений по убыванию."""
    return list(range(length, 0, -1))  # Создание списка от length до 1 включительно

def random_list(length):
    """Генерирует случайный список целых значений."""
    return [random.randint(1, 100) for _ in range(length)]  # Создание списка случайных чисел от 1 до 100 длины length

def almost_sorted_list(length):
    """Генерирует почти отсортированный список целых значений по возрастанию."""
    lst = list(range(1, length + 1))  # Создание списка от 1 до length включительно
    for _ in range(length // 10):  # Перемешивание 10% элементов списка
        index1 = random.randint(0, length - 1)  # Выбор случайного индекса
        index2 = random.randint(0, length - 1)  # Выбор еще одного случайного индекса
        lst[index1], lst[index2] = lst[index2], lst[index1]  # Обмен значениями между элементами по выбранным индексам
    return lst

# Примеры использования:
print("Отсортированный список:", sorted_list(10))
print("Отсортированный в обратном порядке список:", reverse_sorted_list(10))
print("Случайный список:", random_list(10))
print("Почти отсортированный по возрастанию список:", almost_sorted_list(10))
```

11) Реализовать функцию тестирования алгоритма сортировки. Процедура принимает на вход алгоритм сортировки и алгоритм порождения списка f_gen(list_len), генерирующий список заданной длины, количество повтороений теста и список длин использующихся для тестирования. Для каждой длины списка функция тестирования возвращает минимальное, среднее и наибольшее количество сравнений и перестановок.


```python
import random
import statistics

def test_sorting_algorithm(sort_algorithm, list_generator, repetitions, list_lengths):
    results = []  # Создаем список для хранения результатов тестирования

    for length in list_lengths:  # Проходим по каждой длине списка из списка list_lengths
        comparisons = []  # Создаем список для хранения количества сравнений
        permutations = []  # Создаем список для хранения количества перестановок

        for _ in range(repetitions):  # Повторяем тестирование заданное количество раз
            test_list = list_generator(length)  # Генерируем список заданной длины
            comparisons_count, permutations_count = sort_algorithm(test_list)  # Применяем алгоритм сортировки к списку и получаем количество сравнений и перестановок
            comparisons.append(comparisons_count)  # Добавляем количество сравнений в список comparisons
            permutations.append(permutations_count)  # Добавляем количество перестановок в список permutations

        # Находим минимальное, максимальное и среднее количество сравнений
        min_comparisons = min(comparisons)
        max_comparisons = max(comparisons)
        avg_comparisons = statistics.mean(comparisons)

        # Находим минимальное, максимальное и среднее количество перестановок
        min_permutations = min(permutations)
        max_permutations = max(permutations)
        avg_permutations = statistics.mean(permutations)

        # Добавляем результаты тестирования в список results
        results.append({
            'list_length': length,
            'min_comparisons': min_comparisons,
            'max_comparisons': max_comparisons,
            'avg_comparisons': avg_comparisons,
            'min_permutations': min_permutations,
            'max_permutations': max_permutations,
            'avg_permutations': avg_permutations
        })

    return results  # Возвращаем список результатов тестирования

# Пример использования функции test_sorting_algorithm:

# Пример алгоритма сортировки (пузырьковая сортировка)
def bubble_sort(arr):
    comparisons = 0  # Инициализируем счетчик сравнений
    permutations = 0  # Инициализируем счетчик перестановок
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            comparisons += 1  # Увеличиваем счетчик сравнений при каждом сравнении элементов
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                permutations += 1  # Увеличиваем счетчик перестановок при каждой перестановке элементов
    return comparisons, permutations  # Возвращаем количество сравнений и перестановок

# Пример функции генерации случайного списка
def generate_random_list(length):
    return [random.randint(0, 1000) for _ in range(length)]

repetitions = 10  # Количество повторений теста
list_lengths = [10, 50, 100]  # Список длин используемых для тестирования

# Выполняем тестирование и выводим результаты
results = test_sorting_algorithm(bubble_sort, generate_random_list, repetitions, list_lengths)
for result in results:
    print(f"List length: {result['list_length']}, Min comparisons: {result['min_comparisons']}, Max comparisons: {result['max_comparisons']}, Avg comparisons: {result['avg_comparisons']}, Min permutations: {result['min_permutations']}, Max permutations: {result['max_permutations']}, Avg permutations: {result['avg_permutations']}")
```

12) Реализовать модификацию сортировки пузырьком: "камешек-пузырек" - с чередующимися проходами, приводящими к "всплытию" самого большого значения списка и "погружению" самого маленького значения из неотсортированной части списка.


```python
def stone_bubble_sort(arr):
    n = len(arr)  # Получаем длину списка
    for i in range(n):  # Проходим по списку от 0 до n-1
        # В каждой итерации четного прохода (всплытие)
        if i % 2 == 0:  # Проверяем, четное ли текущее значение i
            for j in range(0, n - i - 1, 2):  # Проходим по списку с шагом 2
                if arr[j] > arr[j + 1]:  # Проверяем, нужно ли обменять элементы
                    arr[j], arr[j + 1] = arr[j + 1], arr[j]  # Обмениваем элементы
        # В каждой итерации нечетного прохода (погружение)
        else:  # Если текущее значение i нечетное
            for j in range(1, n - i - 1, 2):  # Проходим по списку с шагом 2, начиная с 1
                if arr[j] > arr[j + 1]:  # Проверяем, нужно ли обменять элементы
                    arr[j], arr[j + 1] = arr[j + 1], arr[j]  # Обмениваем элементы
    return arr  # Возвращаем отсортированный список

# Пример использования:
arr = [64, 34, 25, 12, 22, 11, 90]  # Задаем список для сортировки
sorted_arr = stone_bubble_sort(arr)  # Вызываем функцию сортировки
print("Отсортированный массив:", sorted_arr)  # Выводим отсортированный список
```

15) Реализовать модификацию сортировки выбором на основе выбора из подсписков (описанного в лекции). Протестировать алгоритм.


```python
def selection_sort_modified(arr):
    # Внутренняя функция для поиска индекса минимального элемента в подсписке
    def find_min_index(arr, start):
        min_index = start
        # Проходим по всем элементам после стартового индекса и ищем минимум
        for i in range(start + 1, len(arr)):
            if arr[i] < arr[min_index]:
                min_index = i
        return min_index

    # Основной цикл сортировки выбором
    for i in range(0, len(arr) - 1):
        # Находим индекс минимального элемента в текущем подсписке
        min_index = find_min_index(arr, i)
        # Если индекс минимального элемента не равен текущему индексу, меняем их местами
        if min_index != i:
            arr[i], arr[min_index] = arr[min_index], arr[i]

# Функция для тестирования модифицированной сортировки выбором
def test_selection_sort_modified():
    # Тестовые случаи
    test_cases = [
        [],
        [1],
        [3, 2, 1],
        [5, 2, 7, 3, 9, 1],
        [1, 2, 3, 4, 5],
        [5, 4, 3, 2, 1]
    ]

    # Проходим по каждому тестовому случаю
    for i, test_case in enumerate(test_cases):
        original = test_case.copy()
        # Сортируем массив с помощью модифицированной сортировки выбором
        selection_sort_modified(test_case)
        print(f"Test case {i+1}:")
        print(f"Original array: {original}")
        print(f"Sorted array: {test_case}")
        print()

# Проверяем, запущен ли скрипт напрямую, и если да, то запускаем тестирование
if __name__ == "__main__":
    test_selection_sort_modified()
```

16) Реализовать быструю сортировку с возможностью подсчета операций. Сравнить производительность алгоритма на случайных и упорядоченных (и почти упорядоченных) данных.


```python
def quicksort(arr, start=0, end=None, count_operations=False):
    """
    Реализация алгоритма быстрой сортировки.

    Параметры:
    - arr: список для сортировки.
    - start: начальный индекс для сортировки (по умолчанию 0).
    - end: конечный индекс для сортировки (по умолчанию None, что означает конец списка).
    - count_operations: флаг для подсчета операций (по умолчанию False).

    Возвращает:
    Ничего. Функция сортирует список arr в порядке возрастания.
    """

    # Если конечный индекс не указан, устанавливаем его на конец списка
    if end is None:
        end = len(arr) - 1

    # Если начальный индекс меньше конечного, выполняем сортировку
    if start < end:
        # Находим индекс опорного элемента после разделения списка
        pivot_index = partition(arr, start, end, count_operations)
        # Рекурсивно сортируем элементы перед опорным и после него
        quicksort(arr, start, pivot_index - 1, count_operations)
        quicksort(arr, pivot_index + 1, end, count_operations)


def partition(arr, start, end, count_operations):
    """
    Вспомогательная функция для быстрой сортировки.
    Разделяет список на две части: элементы меньше опорного и элементы больше опорного.

    Параметры:
    - arr: список для сортировки.
    - start: начальный индекс для разделения.
    - end: конечный индекс для разделения.
    - count_operations: флаг для подсчета операций.

    Возвращает:
    Индекс опорного элемента после разделения списка.
    """

    # Задаем опорный элемент (в данном случае - последний элемент списка)
    pivot = arr[end]
    i = start - 1
    # Проходим по списку и перемещаем элементы меньше опорного влево от него
    for j in range(start, end):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
            # Если включен подсчет операций, увеличиваем счетчик на 1
            if count_operations:
                count_operations[0] += 1  # Увеличиваем количество операций на 1
    # Перемещаем опорный элемент в правильную позицию
    arr[i + 1], arr[end] = arr[end], arr[i + 1]
    if count_operations:
        count_operations[0] += 1  # Увеличиваем количество операций на 1
    return i + 1


def test_quicksort_performance(data, ordered=False):
    """
    Функция для тестирования производительности быстрой сортировки на случайных или упорядоченных данных.

    Параметры:
    - data: список для сортировки.
    - ordered: флаг, указывающий, являются ли данные упорядоченными (по умолчанию False).

    Возвращает:
    Кортеж (время выполнения, количество операций).
    """

    import time
    import random

    # Если данные не упорядочены и ordered=True, сортируем их по возрастанию
    if ordered:
        data.sort()

    # Засекаем время начала выполнения
    start_time = time.time()
    # Создаем счетчик операций
    count_operations = [0]  # Счетчик операций
    # Выполняем быструю сортировку
    quicksort(data, count_operations=count_operations)
    # Засекаем время окончания выполнения
    end_time = time.time()

    # Возвращаем время выполнения и количество операций
    return end_time - start_time, count_operations[0]


if __name__ == "__main__":
    # Генерируем случайные данные
    random_data = [random.randint(0, 1000) for _ in range(1000)]

    # Тестируем производительность на случайных данных
    random_time, random_operations = test_quicksort_performance(random_data.copy())
    print("Quicksort performance on random data:")
    print("Time:", random_time, "seconds")
    print("Number of operations:", random_operations)

    # Тестируем производительность на упорядоченных данных
    ordered_time, ordered_operations = test_quicksort_performance(random_data.copy(), ordered=True)
    print("\nQuicksort performance on ordered data:")
    print("Time:", ordered_time, "seconds")
    print("Number of operations:", ordered_operations)
```

17) Улучшить работу быстрой сортировки на упорядоченных и почти упорядоченных за счет изменения алгоритма выбора элемента для разделения массива.


```python
def partition(arr, low, high):
    # Функция partition используется для разделения массива на две части вокруг опорного элемента
    # и возвращает индекс опорного элемента после разделения.
    
    # Выбираем медиану из первого, среднего и последнего элементов в качестве опорного элемента
    mid = (low + high) // 2
    pivot_index = high
    if arr[low] < arr[mid]:
        if arr[mid] < arr[high]:
            pivot_index = mid
    elif arr[low] < arr[high]:
        pivot_index = low

    pivot = arr[pivot_index]

    # Переставляем элементы так, чтобы все элементы слева от pivot были меньше или равны ему,
    # а все элементы справа - больше него
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1


def quick_sort(arr, low, high):
    # Функция quick_sort выполняет быструю сортировку для заданного диапазона индексов массива
    
    if low < high:
        # Если диапазон содержит более одного элемента, производим разделение и рекурсивно сортируем
        # каждую половину массива
        
        # Разбиваем массив и рекурсивно сортируем каждую его половину
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)


# Пример использования:
arr = [10, 7, 8, 9, 1, 5]
n = len(arr)
quick_sort(arr, 0, n - 1)
print("Отсортированный массив:", arr)
```

18) Реализовать эффективный алгоритм получения из большого количества отсортированных списков одного отсортированного списка.


```python
import heapq  # Импортируем модуль heapq для использования кучи

def merge_sorted_lists(sorted_lists):
    merged_list = []  # Инициализируем пустой список для объединенного результата
    heap = []  # Инициализируем пустую кучу

    # Инициализация кучи
    for i, sublist in enumerate(sorted_lists):  # Перебираем отсортированные списки и их индексы
        if sublist:  # Если список не пустой
            heapq.heappush(heap, (sublist[0], i, 0))  # Добавляем в кучу кортеж (значение, индекс_списка, индекс_значения)

    while heap:  # Пока куча не пуста
        val, list_idx, idx = heapq.heappop(heap)  # Извлекаем наименьший элемент из кучи
        merged_list.append(val)  # Добавляем извлеченное значение в объединенный список

        if idx + 1 < len(sorted_lists[list_idx]):  # Если в списке еще есть элементы
            next_val = sorted_lists[list_idx][idx + 1]  # Получаем следующее значение из того же списка
            heapq.heappush(heap, (next_val, list_idx, idx + 1))  # Добавляем его в кучу

    return merged_list  # Возвращаем объединенный список

# Пример использования
sorted_lists = [
    [1, 3, 5],
    [2, 4, 6],
    [0, 7, 8]
]
result = merge_sorted_lists(sorted_lists)
print(result)
```

19) Модифицировать последовательность длин шагов в сортировке Шелла.
На 2[𝑛/2𝑘+1]+1,т.е. при первой сортировке будут использоваться шаги: 2[n/4]+1, 2[n/8]+1, ... , 3, 1 .

```python
def shell_sort(arr):
    n = len(arr)  # Получаем длину массива
    gap = n // 2  # Инициализируем шаг как половину длины массива
    while gap > 0:  # Пока шаг больше нуля
        for i in range(gap, n):  # Начиная с элемента с индексом gap и до конца массива
            temp = arr[i]  # Сохраняем текущий элемент
            j = i
            while j >= gap and arr[j - gap] > temp:  # Пока не достигнут начало массива и текущий элемент меньше предыдущего на gap
                arr[j] = arr[j - gap]  # Перемещаем элемент вперед на gap позиций
                j -= gap  # Уменьшаем индекс на шаг
            arr[j] = temp  # Вставляем сохраненный элемент на правильную позицию
        gap = (gap * 2) // 2 + 1  # Модифицированная формула для расчета шага
    return arr  # Возвращаем отсортированный массив

# Пример использования:
arr = [12, 34, 54, 2, 3]
print("Исходный массив:", arr)
sorted_arr = shell_sort(arr)
print("Отсортированный массив:", sorted_arr)
```
