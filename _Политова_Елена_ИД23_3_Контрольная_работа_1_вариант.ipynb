1. Создайте класс Length (Длина), имеющий свойства

• value (значение),

• unit (единица измерения).

При изменении единицы измерения значение должно соответственно меняться. Например, при переходе от сантиметров к метрам значение должно уменьшаться в 100 раз. Допустимые значения свойства unit: ‘см’, ‘м’, ‘км’. Организуйте эту проверку. Продемонстрируйте работу с классом.

```python
# Решение 1.

class Length:
    def __init__(self, value=0.0, unit='м'):
        self.__value = value
        self.__unit = unit
    
    def get_value(self):
        return self.__value
    
    def get_unit(self):
        return self.__unit
    
    def set_value(self, value):
        self.__value = value
    
    def set_unit(self, unit):
        if unit not in ['см', 'м', 'км']:
            raise ValueError("Unit must be 'см', 'м', or 'км'")
        
        # Конвертируем текущее значение в новую единицу измерения
        if self.__unit == 'см':
            if unit == 'м':
                self.__value /= 100
            elif unit == 'км':
                self.__value /= 100000
        elif self.__unit == 'м':
            if unit == 'см':
                self.__value *= 100
            elif unit == 'км':
                self.__value /= 1000
        elif self.__unit == 'км':
            if unit == 'см':
                self.__value *= 100000
            elif unit == 'м':
                self.__value *= 1000
        
        self.__unit = unit

# Создаем объект класса Length с начальными значениями
length = Length(1, 'м')  # Значение 1 метр

# Выводим текущее значение и единицу измерения
print("Текущее значение:", length.get_value(), length.get_unit())

# Изменяем единицу измерения на сантиметры
length.set_unit('см')
print("Значение в сантиметрах:", length.get_value(), length.get_unit())

# Изменяем единицу измерения на километры
length.set_unit('км')
print("Значение в километрах:", length.get_value(), length.get_unit())
```

2. Создайте класс Ведомость, имеющий
атрибут класса:

• список_дисциплин (значением является список с названиями дисциплин);

свойства:

• дисциплина (при задании значения проверять наличие дисциплины в атрибуте список_дисциплин),

• группа;

методы:

• put – добавляет в ведомость информацию об оценке студента (фамилия, оценка – параметры метода). Для хранения данных внутри класса используйте словарь, в котором ключом является фамилия студента. Возможные оценки – «отлично», «хорошо», «удовл.», «неудовл.», «н/я»;

• get – возвращает оценку, полученную студентом (фамилия студента – параметр метода);

• change – изменяет оценку, полученную студентом (фамилия студента и новая оценка – параметры метода);

• del – удаляет информацию о студенте из ведомости (фамилия студента – параметр метода);

• result – возвращает кортеж из 5 чисел (количество каждого вида оценок в ведомости);

• __init__ – конструктор;

• __str__ – возвращает строку, содержащую заголовки (название экзамена, группа) и результаты экзамена в виде таблицы;

• count – возвращает количество студентов в ведомости;

• names – возвращает список фамилий, имеющихся в ведомости.

Продемонстрируйте работу с классом. 


```python
# Решение 2.


class Ведомость:
    # Атрибут класса - список дисциплин
    список_дисциплин = ['Алгебра', 'ТОД', 'Экономика', 'ОВС', 'История']

    # Конструктор класса
    def __init__(self, дисциплина, группа):
        # Проверка наличия дисциплины в списке допустимых дисциплин
        if дисциплина in self.список_дисциплин:
            self.дисциплина = дисциплина
        else:
            raise ValueError("Недопустимая дисциплина")
        self.группа = группа
        # Словарь для хранения оценок студентов
        self.оценки = {}

    # Метод для добавления оценки студента
    def put(self, фамилия, оценка):
        self.оценки[фамилия] = оценка

    # Метод для получения оценки студента
    def get(self, фамилия):
        return self.оценки.get(фамилия)

    # Метод для изменения оценки студента
    def change(self, фамилия, новая_оценка):
        self.оценки[фамилия] = новая_оценка

    # Метод для удаления информации о студенте
    def __del__(self, фамилия):
        del self.оценки[фамилия]

    # Метод для получения количества каждой оценки
    def result(self):
        отлично = self.оценки.count('отлично')
        хорошо = self.оценки.count('хорошо')
        удовл = self.оценки.count('удовл.')
        неудовл = self.оценки.count('неудовл.')
        ня = self.оценки.count('н/я')
        return (отлично, хорошо, удовл, неудовл, ня)

    # Метод для возврата строкового представления объекта
    def __str__(self):
        результат = f'Результаты экзамена по дисциплине: {self.дисциплина}, Группа: {self.группа}\n'
        for студент, оценка in self.оценки.items():
            результат += f'{студент}: {оценка}\n'
        return результат

    # Метод для подсчета количества студентов в ведомости
    def count(self):
        return len(self.оценки)

    # Метод для получения списка фамилий студентов
    def names(self):
        return list(self.оценки.keys())

# Пример использования класса
ведомость = Ведомость('ТОД', 'ИД23-3')
ведомость.put('Политова', 'отлично')
ведомость.put('Антонова', 'хорошо')
ведомость.put('Смирнов', 'удовл.')
print(ведомость)
print("Количество студентов в ведомости:", ведомость.count())
print("Фамилии студентов в ведомости:", ведомость.names())
```

3. Создайте класс Заказ(Order), у которого есть свойства код_товара(code), цена(price), количество(count) и методы __init__ и __str__.

Создайте 2 класса-потомка: Опт(Opt) и Розница(Retail).
В этих классах создайте методы __init__, __str__.и сумма_заказа(summa), позволяющий узнать стоимость заказа.
Для опта стоимость единицы товара составляет 95% от цены, а при покупке более 500 штук – 90% цены.
В розницу стоимость единицы товара составляет 100% цены. Стоимость заказа равна произведению цены на количество. 

Продемонстрируйте работу с классами, создав необходимые объекты и обратившись к их свойствам и методам


```python
# Решение 3.


# Определение базового класса Заказ
class Order:
    def __init__(self, code, price, count):
        # Инициализация свойств: код товара, цена и количество
        self.code = code
        self.price = price
        self.count = count
    
    def __str__(self):
        # Возвращает строковое представление объекта
        return f'Код: {self.code}, Цена: {self.price}, Количество: {self.count}'

# Определение класса Опт, наследующего класс Заказ
class Opt(Order):
    def __init__(self, code, price, count):
        # Вызов конструктора базового класса
        super().__init__(code, price, count)
    
    def __str__(self):
        # Возвращает строковое представление объекта оптового заказа
        return f'[Опт] {super().__str__()}'

    def summa(self):
        # Расчет стоимости оптового заказа
        if self.count > 500:
            # Скидка 10% при покупке более 500 штук
            return self.count * self.price * 0.9
        else:
            # Скидка 5%
            return self.count * self.price * 0.95

# Определение класса Розница, наследующего класс Заказ
class Retail(Order):
    def __init__(self, code, price, count):
        # Вызов конструктора базового класса
        super().__init__(code, price, count)
    
    def __str__(self):
        # Возвращает строковое представление объекта розничного заказа
        return f'[Розница] {super().__str__()}'

    def summa(self):
        # Расчет стоимости розничного заказа (без скидки)
        return self.count * self.price

# Создание объекта оптового заказа
opt_order = Opt('001', 100, 600)
# Создание объекта розничного заказа
retail_order = Retail('002', 150, 2)

# Вывод информации о заказах и их стоимости
print(opt_order)  # Вывод информации об оптовом заказе
print('Сумма оптового заказа:', opt_order.summa())  # Стоимость оптового заказа

print(retail_order)  # Вывод информации о розничном заказе
print('Сумма розничного заказа:', retail_order.summa())  # Стоимость розничного заказа 
```

4. Написать функцию my_func_1(lst), которая переворачивает список вложенных списков с произвольным уровнем вложенности.
Любой элемент, не являющийся списком, не должен быть изменен.

Примеры, демонстрирующие ожидаемое поведение функции:

assert my_func_1(42) == 42

assert my_func_1([]) == []

assert my_func_1([1, 2]) == [2, 1]

assert my_func_1((1, 2)) == (1, 2)

assert my_func_1([[1, 2], [3, [4, 5]], 6]) == [6, [[5, 4], 3], [2, 1]]

assert my_func_1(['123', '456', '789', {10: [11, 12], 13: 14}, (15, 16)]) == [(15, 16), {10: [11, 12], 13: 14}, '789', '456', '123']

assert my_func_1([[[[[1, 2]]]]]) == [[[[[2, 1]]]]]

```python
# Решение 4.

def my_func_1(lst):
    # Проверяем, является ли входной параметр списком
    if isinstance(lst, list):
        # Инвертируем список и применяем функцию к каждому его элементу
        return [my_func_1(item) for item in reversed(lst)]
    else:
        # Возвращаем элемент без изменений, если он не список
        return lst

# Проверяем функцию на предложенных примерах: 
test_cases = [
    42,
    [],
    [1, 2],
    (1, 2),
    [[1, 2], [3, [4, 5]], 6],
    ['123', '456', '789', {10: [11, 12], 13: 14}, (15, 16)],
    [[[[[1, 2]]]]]
]

results = [my_func_1(case) for case in test_cases]
results

```
