1) Создать скрипт для анализа тестовых файлов - сбора статистики упоминания слов. Аргументом командной строки является имя файла, который нужно проанализировать. Второй аргумент - количество слов, которые нужно выводить на экран (если аргумент не указан, то выводим статистику по 100 словам, статистика выводится в порядке убывания частоты слов, по каждому слову на экран выводится строка в формате "<слово> <частота>"). Если второй аргумент не целое число, то его рассматриваем как имя файла, в который в формате CSV с заголовком сохраняем статистику слов (Первый столбец - слово, второй столбце - частота).

```python
import sys
import csv
from collections import Counter
import string

def analyze_file(filename, num_words=100):
    # Читаем файл и разбиваем его содержимое на слова
    with open(filename, 'r', encoding='utf-8') as file:
        words = file.read().lower().split()

    # Удаляем пунктуацию из слов и создаем счетчик
    translator = str.maketrans('', '', string.punctuation)
    words = [word.translate(translator) for word in words]
    word_count = Counter(words)

    # Сортируем слова по частоте и выводим результат
    sorted_word_count = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
    
    if isinstance(num_words, int):
        sorted_word_count = sorted_word_count[:num_words]
        for word, count in sorted_word_count:
            print(f"{word} {count}")
    else:
        # Если второй аргумент не целое число, считаем его именем файла CSV
        filename_csv = num_words
        with open(filename_csv, 'w', newline='', encoding='utf-8') as csvfile:
            csv_writer = csv.writer(csvfile)
            csv_writer.writerow(['Слово', 'Частота'])
            csv_writer.writerows(sorted_word_count)

if __name__ == "__main__":
    if len(sys.argv) < 2:
        print("Usage: python script.py <filename> [<num_words>]")
    else:
        filename = sys.argv[1]
        if len(sys.argv) >= 3:
            num_words = sys.argv[2]
            if num_words.isdigit():
                num_words = int(num_words)
            else:
                # Если второй аргумент не целое число, считаем его именем файла CSV
                num_words = 100
        else:
            num_words = 100

        analyze_file(filename, num_words)
```

2) Создать скрипт, которому в командной строке передаются имена файлов. В файлах в формате CSV с заголовком содержится информация о частоте слов: первый столбец - слово, второй столбце - частота. Выполнить агрегацию информации о частоте слов из всех переданных файлов (т.е. рассчитать суммарную частоту для всех упомянутых слов). Результат сохранить в файл с именем, которое введет пользователь в ответ на соответствующий запрос.
2.1) Скрипт должен адекватно реагировать на ситуацию отсутствия файлов и другие проблемы с вводом/выводом.
2.2) Оформить модуль с функциями, которые используются в задании 1 и 2.



```python
import csv  # Импорт модуля для работы с CSV файлами
import os   # Импорт модуля для работы с операционной системой

# Функция для агрегации информации о частоте слов из всех переданных файлов
def aggregate_word_frequencies(files):
    word_freq = {}  # Создание пустого словаря для хранения частот слов
    for file_name in files:  # Для каждого переданного имени файла
        try:
            with open(file_name, 'r') as file:  # Открытие файла на чтение
                reader = csv.reader(file)  # Создание объекта чтения CSV
                next(reader)  # Пропуск заголовка
                for row in reader:  # Для каждой строки в файле
                    word, freq = row  # Разделение строки на слово и частоту
                    word_freq[word] = word_freq.get(word, 0) + int(freq)  # Обновление частоты слова в словаре
        except IOError:  # Обработка ошибок ввода/вывода
            print(f"Error: Unable to read file '{file_name}'")  # Вывод сообщения об ошибке
    return word_freq  # Возврат словаря с агрегированными частотами слов

# Функция для сохранения информации о частоте слов в файл
def save_word_frequencies_to_file(word_freq, output_file):
    try:
        with open(output_file, 'w') as file:  # Открытие файла на запись
            writer = csv.writer(file)  # Создание объекта записи CSV
            writer.writerow(['Word', 'Frequency'])  # Запись заголовка
            for word, freq in word_freq.items():  # Для каждой пары слова и его частоты
                writer.writerow([word, freq])  # Запись в файл
        print(f"Word frequencies saved to '{output_file}'")  # Вывод сообщения об успешном сохранении
    except IOError:  # Обработка ошибок ввода/вывода
        print("Error: Unable to write to the output file")  # Вывод сообщения об ошибке

# Основная функция программы
def main():
    files = input("Enter file names separated by spaces: ").split()  # Ввод пользователем имен файлов
    if not files:  # Если пользователь не ввел имена файлов
        print("No files provided.")  # Вывод сообщения об отсутствии файлов
        return  # Выход из функции

    word_freq = aggregate_word_frequencies(files)  # Агрегация частот слов
    if not word_freq:  # Если в файлах не найдены частоты слов
        print("No word frequencies found in the provided files.")  # Вывод сообщения об отсутствии частот
        return  # Выход из функции

    output_file = input("Enter output file name: ")  # Ввод пользователем имени выходного файла
    save_word_frequencies_to_file(word_freq, output_file)  # Сохранение частот слов в файл

# Проверка, что скрипт запущен напрямую (а не импортирован в другой скрипт)
if __name__ == "__main__":
    main()  # Вызов основной функции программы
```

4) Реализовать модуль calculate.py, содержащий набор функций для работы с целыми числами, записанными как на русском языке (например: тридцать восемь), так и в виде простых арифметических выражений (например: 2 + 3).
На базе этого модуля подготовить скрипт calculate.py со следующими возможностями:
1) python claculate.py to_numbers my_file.txt
Вызов приводит к преобразованию всех чисел, написанных на русском языке в числовом виде и сохранению изменений в файл с тем же именем.
python claculate.py to_numbers my_file.txt my_file_res.txt
(результат сохраняется в файл с именем my_file_res.txt)


```python
import re  # Импорт модуля для работы с регулярными выражениями
import pymorphy2  # Импорт библиотеки для морфологического анализа русского текста

# Функция для преобразования слов на русском языке в числовой формат
def to_numbers(text):
    morph = pymorphy2.MorphAnalyzer()  # Создание объекта морфологического анализатора
    words = text.split()  # Разделение текста на слова
    result = []  # Пустой список для хранения результата
    for word in words:
        if re.match(r'^-?\d+$', word):  # Если слово является числом
            result.append(int(word))  # Преобразование слова в целое число и добавление в результат
        else:  # Если слово не число, пытаемся преобразовать его из русской записи
            try:
                parsed = morph.parse(word)[0]  # Морфологический анализ слова
                if 'NUMR' in parsed.tag:  # Если слово является числительным
                    result.append(parsed.normal_form)  # Добавление нормальной формы числительного в результат
                else:  # Если слово не является числительным
                    result.append(word)  # Добавление исходного слова в результат
            except IndexError:  # Если слово не найдено в словаре pymorphy2
                result.append(word)  # Добавление исходного слова в результат
    return result  # Возврат результата

# Функция для обработки содержимого файла
def process_file(input_file, output_file=None):
    with open(input_file, 'r', encoding='utf-8') as file:  # Открытие файла на чтение
        text = file.read()  # Чтение содержимого файла
    converted_text = to_numbers(text)  # Преобразование текста в числовой формат
    if output_file:  # Если указан файл для записи результата
        with open(output_file, 'w', encoding='utf-8') as file:  # Открытие файла на запись
            file.write(' '.join(map(str, converted_text)))  # Запись преобразованного текста в файл
    else:  # Если файл для записи результата не указан
        with open(input_file, 'w', encoding='utf-8') as file:  # Открытие исходного файла на запись
            file.write(' '.join(map(str, converted_text)))  # Запись преобразованного текста в файл

# Проверка, что скрипт запущен напрямую, а не импортирован в другой скрипт
if __name__ == "__main__":
    import sys  # Импорт модуля для работы с системными параметрами и функциями
    if len(sys.argv) < 3:  # Если количество аргументов командной строки меньше 3
        print("Usage: python calculate.py to_numbers input_file [output_file]")  # Вывод сообщения о правильном использовании скрипта
        sys.exit(1)  # Выход из программы с кодом ошибки
    
    input_file = sys.argv[2]  # Получение имени входного файла из аргументов командной строки
    output_file = sys.argv[3] if len(sys.argv) == 4 else None  # Получение имени выходного файла (если указано)

    process_file(input_file, output_file)  # Вызов функции обработки файла
```
